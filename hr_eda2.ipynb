```python
df2=dataf[["Department","Over Time"]]
df2
```

```python
df2['Over Time'] = df2['Over Time'].map({'Yes': 1, 'No': 0})
```

```python
df2
```

```python
#proportion of employees who work overtime based on their department using a pie chart
df_over_time_yes = df2[df2['Over Time'] == 1]
department_counts = df_over_time_yes['Department'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=120)
plt.title('Percentage of Overtime by Department')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#distribution of education levels among employees using a count plot
plt.style.use('dark_background')
a=sns.countplot(x="Education", data=df)
for p in a.patches:
    a.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=11, color='white', xytext=(0, 5),
                textcoords='offset points')
plt.xticks(rotation=30)
plt.yticks(rotation=30)
plt.show()
```

```python
#relationship between distance from home and attrition using a box plot
sns.set(style="whitegrid")
df['CF_current Employee'] = df['CF_current Employee'].map({1: 'Yes', 0: 'No'})
sns.boxplot(x="Distance From Home", y="CF_current Employee", data=df)
plt.show()
```

```python
#average training times last year across different departments
Dep_avg_training_time=df.groupby('Department')['Training Times Last Year'].mean().reset_index()
Dep_avg_training_time
```

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(x="Department", y="Training Times Last Year", data=Dep_avg_training_time, color="lightgreen", palette="prism")
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%.2f' % p.get_height(), 
            fontsize=12, color='black', ha='center', va='bottom')

plt.xlabel('Department')
plt.ylabel('Training Time in Years')
plt.title('Average Training Times Last Year Across Different Departments')
plt.xticks(rotation=45) 
plt.yticks(rotation=45) 
plt.tight_layout()  
plt.show()
```
